```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#line plot
x =[10,15,14,12]
y =[15,10,20,18]


plt.figure(figsize=(3,3))
plt.plot(x, y, "^k--",color="y")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("value of X and Y")
```

```python
x =[10,15,14,12]
y =[15,10,20,18]


plt.figure(figsize=(3,3))
plt.plot(x, y, marker="*",color="y")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("value of X and Y")
```

```python
x =[10,15,14,12]
y =[15,10,20,18]


plt.figure(figsize=(3,3))
sns.lineplot(x=x, y=y,color="y")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("value of X and Y")
```
