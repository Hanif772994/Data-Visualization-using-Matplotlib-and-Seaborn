```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
y=[10,20,34,40]
x=["apple", "samsung","redmi","nothing"]
colo=["r","g","b","y"]
plt.bar(x,y, color=colo, width= .3,)
plt.xlabel("name of mobile brand")
plt.grid()
```

```python
y=[10,20,34,40]
x=["apple", "samsung","redmi","nothing"]
colo=["r","g","b","y"]
sns.barplot(x=x,y=y, width= .4,palette="Set1")
plt.xlabel("name of mobile brand")
plt.legend(["apple", "samsung","redmi","nothing"])
plt.grid()
```
